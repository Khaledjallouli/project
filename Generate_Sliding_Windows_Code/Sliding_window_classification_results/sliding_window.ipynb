## Option 2: sliding window with shots statistics - missing odds dropped

```python
data_shots = process_csv_data_shots.Dataset('../../data/OldData/matches_all_with_shots.csv')
df_shots =  pd.DataFrame(data_shots.processed_results)
df_shots.to_csv('../../data/data_classification_results/sliding02_shots.csv')
df_shots
```

## Option 3: sliding window with shots statistics and additional accuracy features - missing odds dropped

```python
df_shots_extra = df_shots
df_shots_extra['home_shot_accuracy'] = df_shots_extra['home-shots_on_target']/df_shots_extra['home-shots']
df_shots_extra['home_shot_efficiency'] = df_shots_extra['home-goals']/df_shots_extra['home-shots_on_target']
df_shots_extra['home_opposition_shot_accuracy'] = df_shots_extra['home-opposition_shots_on_target']/df_shots_extra['home-opposition_shots']
df_shots_extra['home_opposition_shot_efficiency'] = df_shots_extra['home-opposition-goals']/df_shots_extra['home-opposition_shots_on_target']

df_shots_extra['away_shot_accuracy'] = df_shots_extra['away-shots_on_target']/df_shots_extra['away-shots']
df_shots_extra['away_shot_efficiency'] = df_shots_extra['away-goals']/df_shots_extra['away-shots_on_target']
df_shots_extra['away_opposition_shot_accuracy'] = df_shots_extra['away-opposition_shots_on_target']/df_shots_extra['away-opposition_shots']
df_shots_extra['away_opposition_shot_efficiency'] = df_shots_extra['away-opposition-goals']/df_shots_extra['away-opposition_shots_on_target']

df_shots_extra.to_csv('../../data/data_classification_results/sliding03_shots_extra.csv')
df_shots_extra
```

```python
import pandas_profiling
profile = df_shots_extra.profile_report(title='Sliding Window with extra features')
profile.to_file(output_file="sliding03_shots_extra_profile.html")
df_shots_extra.profile_report(style={'full_width':True})
```
